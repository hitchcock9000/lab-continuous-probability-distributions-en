```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import seaborn as sns
```

```python
# This is an example of how to generate uniform random numbers
# numpy.random.uniform() creates random numbers where each value has equal probability
# The parameters are: low (minimum value), high (maximum value), size (how many numbers)

# Generate 5 random numbers between 0 and 10
example_uniform = np.random.uniform(low=0, high=10, size=5)
print("Example uniform random numbers:", example_uniform)
```

```python
# Creating a function to generate uniform random numbers
# This function will make it easy to create uniform distributions with different parameters

def generate_uniform(bottom, ceiling, count):
    """
    Generate uniformly distributed random numbers.
    
    Parameters:
    - bottom: the minimum value (lower boundary)
    - ceiling: the maximum value (upper boundary)
    - count: how many random numbers to generate
    
    Returns:
    - An array of random numbers uniformly distributed between bottom and ceiling
    """
    # np.random.uniform generates random numbers where every value has equal chance
    return np.random.uniform(low=bottom, high=ceiling, size=count)

# Test the function with a small example
test_data = generate_uniform(0, 5, 10)
print("Test - 10 random numbers between 0 and 5:")
print(test_data)
```

```python
# Now let's generate the two datasets as requested

# First dataset: 100 numbers between 10 and 15
# This is a narrow range (only 5 units wide)
uniform_data_1 = generate_uniform(bottom=10, ceiling=15, count=100)

# Second dataset: 1000 numbers between 10 and 60
# This is a much wider range (50 units wide)
uniform_data_2 = generate_uniform(bottom=10, ceiling=60, count=1000)

print("Dataset 1 - First 10 values:", uniform_data_1[:10])
print("Dataset 2 - First 10 values:", uniform_data_2[:10])
```

```python
# Creating histograms to visualize the distributions
# A histogram shows how many values fall into different ranges (bins)

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# First histogram: 100 numbers between 10 and 15
ax1.hist(uniform_data_1, bins=10, color='skyblue', edgecolor='black')
ax1.set_title('Uniform Distribution (10-15, n=100)', fontsize=12)
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency (Count)')
ax1.grid(True, alpha=0.3)

# Second histogram: 1000 numbers between 10 and 60
ax2.hist(uniform_data_2, bins=10, color='lightcoral', edgecolor='black')
ax2.set_title('Uniform Distribution (10-60, n=1000)', fontsize=12)
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency (Count)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# How are the two distributions different?

# The main differences are:
# 1. RANGE: The first distribution covers a narrow range (10-15), while the second covers a wide range (10-60)
# 2. SAMPLE SIZE: The first has 100 samples, the second has 1000 samples (10x more data)
# 3. BAR HEIGHTS: The second distribution has taller bars because it has more samples
# 4. SPREAD: Both are uniform (flat), but the second is spread over a much wider area
#
# Despite these differences, both are UNIFORM distributions, meaning:
# - Each value in the range has an equal probability of being selected
# - The bars should be roughly the same height (with some random variation)
```

```python
# Creating a function to generate normally distributed random numbers
# Normal distribution is the famous "bell curve" - most values cluster around the mean

def generate_normal(mean, std, count):
    """
    Generate normally distributed random numbers.
    
    Parameters:
    - mean: the average value (center of the bell curve)
    - std: standard deviation (how spread out the values are)
    - count: how many random numbers to generate
    
    Returns:
    - An array of random numbers following a normal distribution
    """
    # np.random.normal creates a bell curve distribution
    # Most values will be close to the mean, fewer values far from it
    return np.random.normal(loc=mean, scale=std, size=count)

# Test the function
test_normal = generate_normal(mean=0, std=1, count=10)
print("Test - 10 numbers from a normal distribution (mean=0, std=1):")
print(test_normal)
```

```python
# Generate first dataset: mean=10, standard deviation=1
# With std=1, most values will be very close to 10 (between 9 and 11)
normal_data_1 = generate_normal(mean=10, std=1, count=1000)

print("Dataset 1 (mean=10, std=1):")
print(f"  First 10 values: {normal_data_1[:10]}")
print(f"  Actual mean: {normal_data_1.mean():.2f}")
print(f"  Actual std: {normal_data_1.std():.2f}")
```

```python
# Generate second dataset: mean=10, standard deviation=50
# With std=50, values will be much more spread out (could range from -100 to +120)
normal_data_2 = generate_normal(mean=10, std=50, count=1000)

print("Dataset 2 (mean=10, std=50):")
print(f"  First 10 values: {normal_data_2[:10]}")
print(f"  Actual mean: {normal_data_2.mean():.2f}")
print(f"  Actual std: {normal_data_2.std():.2f}")
```

```python
# Creating histograms to visualize the normal distributions

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# First histogram: mean=10, std=1 (narrow bell curve)
ax1.hist(normal_data_1, bins=30, color='lightgreen', edgecolor='black')
ax1.set_title('Normal Distribution (mean=10, std=1, n=1000)', fontsize=12)
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency (Count)')
ax1.grid(True, alpha=0.3)

# Second histogram: mean=10, std=50 (wide bell curve)
ax2.hist(normal_data_2, bins=30, color='plum', edgecolor='black')
ax2.set_title('Normal Distribution (mean=10, std=50, n=1000)', fontsize=12)
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency (Count)')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Creating a function to generate exponentially distributed random numbers
# Exponential distribution is used for modeling time between events
# (like time between customer arrivals, or time until a machine breaks)

def generate_exponential(mean, count):
    """
    Generate exponentially distributed random numbers.
    
    Parameters:
    - mean: the average value
    - count: how many random numbers to generate
    
    Returns:
    - An array of random numbers following an exponential distribution
    """
    # The scale parameter in numpy is the same as the mean
    # Exponential distribution: many small values, few large values
    return np.random.exponential(scale=mean, size=count)

# Generate the two datasets as requested
# Small sample: only 10 numbers
exp_data_1 = generate_exponential(mean=10, count=10)

# Large sample: 100 numbers
exp_data_2 = generate_exponential(mean=10, count=100)

print("Small sample (n=10):", exp_data_1)
print("\nLarge sample (n=100) - first 10 values:", exp_data_2[:10])

# Plot both distributions
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Histogram for 10 samples
ax1.hist(exp_data_1, bins=100, color='purple', edgecolor='black', alpha=0.7)
ax1.set_title('Exponential Distribution (mean=10, n=10)', fontsize=12)
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')
ax1.grid(True, alpha=0.3)

# Histogram for 100 samples
ax2.hist(exp_data_2, bins=100, color='teal', edgecolor='black', alpha=0.7)
ax2.set_title('Exponential Distribution (mean=10, n=100)', fontsize=12)
ax2.set_xlabel('Value')
ax2.set_ylabel('Frequency')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Bank waiting time problem
# We need to use the exponential distribution from scipy.stats
from scipy.stats import expon

# Given information:
# - Mean time in bank = 10 minutes
# - This means  (lambda) = 1/10 = 0.1
# - In scipy, we use 'scale' parameter which equals the mean (10)

mean_time = 10  # average time in minutes

# Question: What is P(X < 15)? 
# This means: probability that someone spends LESS than 15 minutes

# We use the CDF (Cumulative Distribution Function)
# CDF gives us the probability of being less than or equal to a value
prob_less_than_15 = expon.cdf(15, scale=mean_time)

print("Probability that a customer spends less than 15 minutes in the bank:")
print(f"P(X < 15) = {prob_less_than_15:.4f}")
print(f"As a percentage: {prob_less_than_15 * 100:.2f}%")
print("\nInterpretation: About 78% of customers will spend less than 15 minutes in the bank.")
```

```python
# Question: What is the probability that a customer spends MORE than 15 minutes?
#
# This is P(X > 15)
# We can calculate this as: 1 - P(X < 15)
# Because the total probability must equal 1

prob_more_than_15 = 1 - prob_less_than_15

# Alternative way: using the survival function (sf)
# The survival function directly gives us P(X > x)
prob_more_than_15_alt = expon.sf(15, scale=mean_time)

print("Probability that a customer spends more than 15 minutes in the bank:")
print(f"P(X > 15) = {prob_more_than_15:.4f}")
print(f"As a percentage: {prob_more_than_15 * 100:.2f}%")
print(f"\nVerification using survival function: {prob_more_than_15_alt:.4f}")
print("\nInterpretation: About 22% of customers will spend more than 15 minutes in the bank.")
print("\nNote: P(X < 15) + P(X > 15) = 1.0 ")
```

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank?
